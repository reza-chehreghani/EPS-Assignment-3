```python
import numpy as np, random
import matplotlib.pyplot as plt

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
set_seed(810109203)

import pandas as pd
df = pd.read_csv('FIFA2020.csv', encoding = "ISO-8859-1")

mean_pace = df['pace'].mean()
df['pace'] = df['pace'].replace(np.nan, mean_pace)

mean_dribbling = df['dribbling'].mean()
df['dribbling'] = df['dribbling'].replace(np.nan, mean_dribbling)

import seaborn as sns
sns.set_style("whitegrid")
sns.boxplot(x='age', data=df)

describe_age = df['age'].describe()
print(describe_age[3:].to_string())
```

```python
n = 100
set_seed(810109203)
sample_weight = np.random.choice(df['weight'].to_numpy(), n, False)
mean_sample_weight = np.mean(sample_weight)
var_sample_weight = np.var(sample_weight)
std_sample_weight = np.std(sample_weight)
print('mean sample weight:', mean_sample_weight)
print('var sample weight:', var_sample_weight)
print('std sample weight:', std_sample_weight)

sample_norm = np.random.normal(mean_sample_weight, std_sample_weight, n)

sorted_sample_weight = np.sort(sample_weight)
sorted_sample_norm = np.sort(sample_norm)

plt.scatter(sorted_sample_norm, sorted_sample_weight)

x = (sorted_sample_norm[0], sorted_sample_norm[-1])
plt.plot(x, x)    #  y = x

plt.xlabel('Normal theoretical quantiles')
plt.ylabel('Data quantiles')
plt.show()
```

```python
import scipy.stats as stats
statistic, p_value = stats.shapiro(sample_weight)

print('statistic:', statistic)
print('p_value:', p_value)
```

```python
var = lam
std = np.sqrt(var)

n = 5
set_seed(810109203)
sample_poisson = np.random.poisson(lam, n)
sample_norm = np.random.normal(lam, std, n)

sorted_sample_poisson = np.sort(sample_poisson)
sorted_sample_norm = np.sort(sample_norm)

plt.scatter(sorted_sample_norm, sorted_sample_poisson)

x = (sorted_sample_norm[0], sorted_sample_norm[-1])
plt.plot(x, x)    #  y = x

plt.xlabel('Normal theoretical quantiles')
plt.ylabel('Poisson theoretical quantiles')
plt.show()

statistic, p_value = stats.shapiro(sample_poisson)
print('p_value:', p_value)
```
