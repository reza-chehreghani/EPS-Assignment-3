```python
import numpy as np, random
import matplotlib.pyplot as plt

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
set_seed(810109203)

def calc_linear_regression(x_points, y_points):
    mean_x = np.mean(x_points)
    mean_y = np.mean(y_points)

    diff_x = x_points - mean_x
    squared_diff_x = diff_x * diff_x

    # Y = a + bX
    b = np.sum(y_points * diff_x) / np.sum(squared_diff_x)
    a = mean_y - b * mean_x
    return a, b

def predeict_by_linear_regression(a, b, x_points):
    y = a + b * x_points
    return y

def calc_coefficient_of_determination(x_points, y_points, a, b):
    linear_regression_predict = predeict_by_linear_regression(a, b, x_points)

    # diff y_points and linear_regression_predict
    residual = y_points - linear_regression_predict
    squared_residual = residual * residual

    # sum of squares of residuals
    SS_res = np.sum(squared_residual)

    mean_y = np.mean(y_points)
    diff_y = y_points - mean_y
    squared_diff_y = diff_y * diff_y

    # total sum of squares
    SS_tot = np.sum(squared_diff_y)

    coefficient_of_determination = 1 - SS_res / SS_tot
    return coefficient_of_determination


x_points = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5])
y_points = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5])

a, b = calc_linear_regression(x_points, y_points)

plt.scatter(x_points, y_points)

x = np.linspace(np.min(x_points), np.max(x_points), 100)
y = predeict_by_linear_regression(a, b, x)
plt.plot(x, y)

coefficient_of_determination = calc_coefficient_of_determination(x_points, y_points, a, b)
plt.title(f'Coefficient of determination = {coefficient_of_determination:.2f}')
plt.show()

print(coefficient_of_determination)
```
